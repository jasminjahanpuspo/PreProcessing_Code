```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import ndimage
from scipy.ndimage import median_filter
```

```python
im = plt.imread("E:\\images (1).jpg")
im.shape
```

```python
plt.imshow(im,cmap='gray')
```

```python
b, g, r = cv2.split(im)
```

```python
def simple_threshold(im, threshold=128):
    return ((im > threshold) * 255).astype("uint8")

thresholds = [50,100,150,200,250]
c_ims = []
fig, axs = plt.subplots(nrows=1, ncols=len(thresholds), figsize=(20,5));
gray_im = b
print(gray_im.shape)
                        
for t, ax in zip(thresholds, axs):
    tmp_im=simple_threshold(gray_im, t)
    ax.imshow(simple_threshold(gray_im, t), cmap='Greys');
    c_ims.append(tmp_im)
    ax.set_title("Threshold: {}".format(t), fontsize=20);
    ax.set_axis_off();
```

```python
print(len(c_ims))
```

```python
plt.imshow(c_ims[3] & c_ims[1] & c_ims[2], cmap='Greys')
```
